WebScraping of Indeed Jobsite

```
##Webscrping of Indeed Jobsite

from bs4 import BeautifulSoup
import pandas as pd
import requests

def extract(search_location, radius, page):
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'} 
  #url = f"https://uk.indeed.com/jobs?q=&l={search_location}&start={page}"
  url = f"https://www.indeed.co.uk/jobs?q=&l={search_location}+Essex&radius={radius}&sort=date&start={page}"
  #url=f"https://uk.indeed.com/jobs?q=&l=Essex&radius=0&start={page}"
  r= requests.get(url,headers)
  soup = BeautifulSoup(r.text, 'html.parser')
  return soup


def transform(soup):
  divs = soup.find_all('div', class_ ='jobsearch-SerpJobCard unifiedRow row result')
  for job in divs:

      job_title_obj = job.find("a").text.strip().replace('\n', '').replace('  ', '')
      #print(job_title_obj)
      
      

      try:
        company_obj = job.find("span", {"class": "company"}).text.replace( '\n', '').replace('  ', '')
        
        
      except:
          company_obj = " "

      try:
        date_obj = job.find("span", {"class": "date"}).text.replace( '\n', '').replace('  ', '')
        
        
      except:
          date_obj = " "
      # Location - span HTML tag
      try:
        loc_obj = job.find("span", class_= 'location accessible-contrast-color-location').text.replace('\n', '').replace('  ', '')
        
        
      except:
        loc_obj = ""

      # Location - div HTML tag


      try:
        salary_obj = job.find("span", class_ =  "salaryText").text.replace('\n', '').replace('  ', '')
        #salary_info = clean_salary(salary_obj)
        #job_features.append(salary_info[0])  # salary amount
        #ob_features.append(salary_info[1])  # salary frequency
      except:
        salary_obj = " "

      ##Persistent_URL
      try:
        # Finds the title tag for each job, its associated anchor tag and extracts the href for persistent job webpage
        persist_url = job.find('h2').find('a').get('href')
        # Webscrape the persistent job webpage for each job
        detailed_url = f'https://www.indeed.co.uk{persist_url}'
        
      except:
        detailed_url =" "

      #Job_Description
      try:
        summary_obj = job.find('div', {'class' : 'summary'}).text.replace( '\n', '').replace('  ', '')
        
        
      except:
        summary_obj = " "

      job = {
           'Job_title': job_title_obj,
           'Employer' : company_obj,
           'date': date_obj,
           'Location'  : loc_obj,
           #'Location_span': loc_obj2,
           'Salary' : salary_obj, 
           'detailed_url': detailed_url,
           'Job_description': summary_obj
            }
      joblist.append(job)
  return
```

```
joblist=[]
search_location = ['Basildon', 'Braintree', 'Brentwood','Benfleet', 'Chelmsford', 'Colchester' ,'Epping', 'Harlow', 'Maldon', 'Rochford', 'Southend-on-Sea', 'Tendring','Thurrock', 'Uttlesford'] 
                 
radius = ['0','5','10', '15', '25' , '50','100']
for j in search_location:
    
    print(f'Loading==============================> ,{j}')
    for r in radius:
        print(f'Getting Radius,{r}')
        for i in range(0, 750, 10):
            print(f'Getting Page, {i}')
            c= extract(j,r,i)
            transform(c)
print(len(joblist))

```
